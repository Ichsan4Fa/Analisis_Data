### Cleaning Data

<h3> Dataset day_df

Tahapan pembersihan meliputi penghapusan duplikasi, mengimputasi missing value, dan mengubah nilai data outlier. Setelah itu, dataset akan dicek kembali jika proses cleaning selesai.

```python
#Mengatasi duplikasi data
day_df.drop_duplicates(inplace=True)
```

```python
#Mengatasi Missing Value
day_df.windspeed.fillna(value=day_df.windspeed.mean(),inplace=True)
day_df.temp.fillna(value=day_df.temp.mean(),inplace=True)
day_df.atemp.fillna(value=day_df.atemp.mean(),inplace=True)
day_df.hum.fillna(value=day_df.hum.mean(),inplace=True)
```

```python
#Mengatasi outlier di kolom cnt
Q1 = (day_df['cnt']).quantile(0.25)
Q3 = (day_df['cnt']).quantile(0.75)
IQR = Q3 - Q1

maximum = Q3 + (1.5*IQR)
minimum = Q1 - (1.5*IQR)

kondisi_lower_than = day_df['cnt'] < minimum
kondisi_more_than = day_df['cnt'] > maximum

day_df.loc[kondisi_more_than, 'cnt'] = maximum
day_df.loc[kondisi_lower_than, 'cnt'] = minimum
```

##RFM Analysis

```python
rfm_df = all_df.groupby(by="instant", as_index=False).agg({
    "dteday_x": "max", #mengambil tanggal order terakhir
    "holiday_x": "nunique",
    "casual_x": "sum"
})
rfm_df.columns = ["index", "max_date", "frequency", "monetary"]
rfm_df.head()
```

```python
#Mencari nilai recency dengan mencari selisih tanggal rfm_df dengan day_df
day_df["dteday"]=pd.to_datetime(day_df['dteday'])
rfm_df["max_date"]=pd.to_datetime(rfm_df['max_date'])
rfm_df["max_date"] = rfm_df["max_date"].dt.date
recent_date = day_df["dteday"].dt.date.max()
rfm_df["recency"] = rfm_df["max_date"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_date", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

```python
rfm_df.sort_values(by="recency", ascending=True).head(5)
```

```python
rfm_df.sort_values(by="frequency", ascending=False).head(5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sn.barplot(y="recency", x="index", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sn.barplot(y="frequency", x="index", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sn.barplot(y="monetary", x="index", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Most Casual Renters Based on RFM Parameters (instant)", fontsize=20)
plt.show()
```
